```python
def box(model,img) :
    results=model(img)
    for i in results :
        boxes = i.boxes  # Boxes object for bbox outputs
        masks = i.masks  # Masks object for segmentation masks outputs
        keypoints = i.keypoints  # Keypoints object for pose outputs
        probs = i.probs  # Probs object for classification outputs
    conf=boxes.conf
    xyxy=boxes.xyxy
    cls=boxes.cls
    names={0:"bir",1:"duz",2:"iki",3:"sag",4:"sol",5:"uc"}
    for i in range(len(cls)) :           
        p1=(int(xyxy[i][0]),int(xyxy[i][1]))
        p2=(int(xyxy[i][2]),int(xyxy[i][3]))
        p3=(int(xyxy[i][0]),int(xyxy[i][3]))
        img=cv2.rectangle(img,p1,p2,(255,0,0),2)
        img=cv2.putText(img,"{} :{:.2f}".format(names[int(cls[i])],float(conf[i])),p3,cv2.FONT_HERSHEY_SIMPLEX ,1,(255,0,0),2)
    return img
```
